## 필요 라이브러리 가져오기

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

# os 별로 폰트를 다르게 설정해 줍니다.
if os.name == "posix":
    # Mac
    sns.set(font="AppleGothic")
elif os.name == "nt":
    # Window
    sns.set(font="Malgun Gothic")
```

```python
# 레티나 설정을 해주면 글씨가 좀 더 선명하게 보입니다.
# 폰트의 주변이 흐릿하게 보이는 것을 방지합니다.
%config InlineBackend.figure_format = 'retina'
```

## 데이터 로드하기

```python
df_raw = pd.read_csv("data/국가_대륙_별_상품군별_온라인쇼핑_해외직접판매액_202001.csv", 
                     encoding="cp949")
df_raw.shape
```

```python
# "국가(대륙)별" 데이터 빈도수 세기
df_raw["국가(대륙)별"].value_counts()
```

```python
# 미국 데이터만 따로 보기
df_raw[df_raw["국가(대륙)별"] == "미국"].head()
```

## 분석과 시각화를 위한 tidy data 만들기
* https://pandas.pydata.org/Pandas_Cheat_Sheet.pdf

```python
# melt() : 행에 있는 데이터를 열로 옮긴다
# id_vars : id값으로 사용할 열을 지정
# variable, value column이 자동 생성된다 -> var_name, value_name을 따로 지정해줄 수 있다

df = df_raw.melt(id_vars=["국가(대륙)별", "상품군별", "판매유형별"], 
                 var_name="기간", value_name="백만원")
df.shape
```

```python
df
```

```python
df.info()
```

## 데이터 전처리
### 기간에서 연도를 분리하기

```python
# melt()한 데이터에 "연도" column을 추가하기

df["연도"] = df["기간"].map(lambda x: x.split()[0])
df.head()
```

```python
# 기간에서 분기 숫자만 가져오기

int("2019 3/4 p)".split()[1].split("/")[0])
```

```python
df["분기"] = df["기간"].map(lambda  x : x.split()[1].split("/")[0])
df.head()
```

### 금액을 수치데이터로 표현하기 위해 데이터 타입 변경하기

```python
# np: numpy, np.nan : NaN

df["백만원"] = df["백만원"].replace("-", pd.np.nan).astype(float)
df
```

### 필요없는 데이터 제거하기

```python
df.info()
```

```python
# 합계 데이터는 따로 구할 수 있기 때문에 전체 데이터에서 제거합니다.

df = df[(df["국가(대륙)별"] != "합계") & 
        (df["상품군별"] != "합계")].copy()
df.info()
```

```python
df.isnull().sum()
```

## K-Beauty 시각화
### 전체 상품군 판매액

```python
df_total = df[df["판매유형별"] == "계"].copy()
df_total
```

```python
sns.lineplot(data=df_total, x="연도", y="백만원")
```

```python
sns.lineplot(data=df_total, x="연도", y="백만원", hue="상품군별")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.relplot(data=df_total, x="연도", y="백만원", 
            hue="상품군별", kind="line", col="상품군별", col_wrap=4)
```

```python
# isin() : return boolean dataframe whether value is contained in
# 화장품 품목만 뺀다
# 의류, 패션 품목도 뺀다

df_sub = df_total[~df_total["상품군별"].isin(["화장품", "의류 및 패션관련 상품"])].copy()
```

```python
sns.relplot(data=df_sub, x="연도", y="백만원", 
            hue="상품군별", kind="line", col="상품군별", col_wrap=4)
```

### 화장품의 온라인쇼핑 해외직접판매액

```python
df_cosmetic = df_total[df_total["상품군별"] == "화장품"].copy()
df_cosmetic["상품군별"].unique()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=df_cosmetic, x="연도", y="백만원", hue="분기")
```

```python
# plt.xticks(rotation=angle) : x축 label의 기울기

plt.figure(figsize=(15, 5))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x="기간", y="백만원", hue="국가(대륙)별")
```

```python
plt.figure(figsize=(15, 5))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic[df_cosmetic["국가(대륙)별"] != "중국"], 
             x="기간", y="백만원", hue="국가(대륙)별")
```

```python
df_sub = df[(df["판매유형별"] != "계") & (df["판매유형별"]) != "면세점"].copy()

plt.figure(figsize=(15, 5))
plt.xticks(rotation=30)
sns.lineplot(data=df_sub, x="기간", y="백만원", hue="판매유형별", ci=None)
```

### 의류 및 패션관련 상품 온라인쇼핑 해외직접판매액

```python
df_fashion = df[(df["상품군별"] == "의류 및 패션관련 상품") & (df["판매유형별"] == "계")].copy()
```

```python
plt.figure(figsize=(15, 5))
plt.title("의류 및 패션관련 상품")
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x="기간", y="백만원", hue="국가(대륙)별", ci=None)
```

```python
df_fashion2 = df[(df["상품군별"] == "의류 및 패션관련 상품") & (df["판매유형별"] != "계")].copy()
```

```python
plt.figure(figsize=(15, 5))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion2, x="기간", y="백만원", hue="판매유형별", ci=None)
```

### 데이터 집계하기

```python
df_fashion["판매유형별"].value_counts()
```

```python
# pivot_table(aggfunc) : default는 mean, sum으로하면 분기별 데이터가 연도 column에 합으로 나타난다

result = df_fashion.pivot_table(
    index="국가(대륙)별", columns="연도", values="백만원",
    aggfunc="sum")
result
```

### 연산결과를 시각적으로 보기

```python
plt.figure(figsize=(15, 7))
sns.heatmap(result, cmap="Blues", annot=True, fmt=".0f")
```
